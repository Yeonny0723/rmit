### 3. Date and Datetime

```python
import datetime

# ###########
# Date object
# ###########

# Get current date
date_object = datetime.date.today()
print(date_object)
print()

# Create a Date object
d = datetime.date(2019, 1, 15) # yyyy, mm, dd
print(d)
print()

# Today's date
today = datetime.date.today()
print("Current year:", today.year)
print("Current month:", today.month)
print("Current day:", today.day)
print()
    
# ###############
# Datetime object
# ##############    
    
#  Get Current Date and Time
datetime_object = datetime.datetime.now()
print(datetime_object)    
print()
  
# Create a datetime object    
datetime_object = datetime.datetime(2019, 1, 15, 23, 55, 59, 342380) # yyyy, mm, dd, hr, minute, sec, microsecond
print(datetime_object)
print()
    
# Create a datetime object - ignore the time
datetime_object = datetime.datetime(2019, 1, 15) # yyyy, mm, dd
print(datetime_object)
print()

# ################
# Date difference
# ################
d1 = datetime.date(2019, 1, 15)
d2 = datetime.date(2019, 1, 1)
num_days = d1 - d2
print('date difference:',num_days)
print()


# ##############################################
# Parsing MySQL date format to python date object
# ##############################################
# MySQL date format: yyyy-mm-dd
# Australian/English date format: dd/mm/yyyy
from dateutil.parser import parse # Need to import parse from the date parser

# Create a python date object by parsing an Australian-formatted date string
# Useful when an Australian date string has been entered in a GUI
# This python data object can be stored in an SQLIte database
# A MYSQL database cannot store a python date object - must be a string in yyyy-mm-dd format!
date_obj = parse("15/1/2019") # parse a string formatted as dd/mm/yyyy
print(date_obj)
print()

# Format a python date object as a MySQL date formatted string (yyyy-mm-dd)
# This string can then be stored in the MySQL database
mysql_date_format = date_obj.strftime('%Y-%m-%d') # format the date as yyyy-mm-dd
print(mysql_date_format)
print()

# Format a python data object as an Australian/English date format dd/mm/yyyy
# This is useful for displaying a python date object read from a database in dd/mm/yy format
oz_date_format = date_obj.strftime('%d/%m/%Y') # format the date as dd/mm/yyyy
print(oz_date_format)
print()
```
